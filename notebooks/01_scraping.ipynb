```python
import sys
import os
import json
import subprocess
import pandas as pd

project_root = os.path.dirname(os.getcwd()) if 'notebooks' in os.getcwd() else os.getcwd()
sys.path.insert(0, project_root)

from src.scraping.models import Product, Review, ScrapedData
from src.scraping.data_manager import DataManager
from config import settings

print(f"✅ Ready!")
print(f"📁 Project: {project_root}")
```

```python
def run_scraper(urls: list, max_reviews: int = 500) -> dict:
    """Run scraper with improved pagination."""
    urls_json = json.dumps(urls)
    python_exe = sys.executable
    runner_path = os.path.join(project_root, "src", "scraping", "runner.py")
    
    print(f"🚀 Running scraper for {len(urls)} URLs")
    print(f"📊 Max {max_reviews} reviews per product")
    print(f"⏱️ Browser akan terbuka...\n")
    
    result = subprocess.run(
        [python_exe, runner_path, "--urls", urls_json, "--max-reviews", str(max_reviews)],
        capture_output=True, text=True, encoding='utf-8', cwd=project_root
    )
    
    if result.stderr:
        print(result.stderr)
    
    if result.stdout.strip():
        try:
            return json.loads(result.stdout)
        except:
            return {"products": [], "reviews": [], "errors": []}
    return {"products": [], "reviews": [], "errors": []}

print("✅ Scraper function ready!")
```

## 2️⃣ Load All URLs (37 total)

```python
with open(os.path.join(project_root, "link.txt"), 'r', encoding='utf-8') as f:
    all_urls = [line.strip() for line in f if line.strip().startswith('http')]

print(f"📋 Total URLs: {len(all_urls)}")

# Categorize
priority_stores = ['ismile-indonesia', 'milano-cell', 'senz88', 'digitechmall', 
                   'ibox-official', 'digimap-official-store', 'goodponsel']
priority_urls = [u for u in all_urls if any(s in u for s in priority_stores)]
regular_urls = [u for u in all_urls if u not in priority_urls]

print(f"  🔥 Priority: {len(priority_urls)}")
print(f"  📦 Regular: {len(regular_urls)}")
```

## 3️⃣ Priority Scraping (High-Review Stores)

⏱️ Fokus pada toko dengan banyak review

```python
print("🔥 PRIORITY SCRAPING")
print(f"Scraping {len(priority_urls)} high-review stores...\n")

priority_result = run_scraper(priority_urls, max_reviews=1000)

print(f"\n{'='*50}")
print(f"📊 PRIORITY RESULTS")
print(f"{'='*50}")
print(f"Products: {len(priority_result['products'])}")
print(f"Reviews: {len(priority_result['reviews'])}")
```

## 4️⃣ Regular Scraping (Other Stores)

```python
print("📦 REGULAR SCRAPING")
print(f"Scraping {len(regular_urls)} regular stores...\n")

regular_result = run_scraper(regular_urls, max_reviews=200)

print(f"\n{'='*50}")
print(f"📊 REGULAR RESULTS")
print(f"{'='*50}")
print(f"Products: {len(regular_result['products'])}")
print(f"Reviews: {len(regular_result['reviews'])}")
```

```python
# Di notebook, cell baru:
test_url = ["https://www.tokopedia.com/milano-cell/fs-milano-apple-iphone-17-256gb-garansi-resmi-1733680775158727703"]
test_result = run_scraper(test_url, max_reviews=1000)
print(f"Milano reviews: {len(test_result['reviews'])}")
```

## 5️⃣ Combine & Save (Fresh - No Dedup with Old Data)

```python
# Combine results
all_products = priority_result.get('products', []) + regular_result.get('products', [])
all_reviews = priority_result.get('reviews', []) + regular_result.get('reviews', [])

print(f"📊 COMBINED RESULTS")
print(f"{'='*50}")
print(f"Total products: {len(all_products)}")
print(f"Total reviews (raw): {len(all_reviews)}")

# Convert to models
scraped_data = ScrapedData()

for p in all_products:
    scraped_data.add_product(Product(
        url=p.get('url', ''),
        name=p.get('name'),
        price=p.get('price'),
        store_name=p.get('store_name'),
        rating=p.get('rating'),
        review_count=p.get('review_count')
    ))

for r in all_reviews:
    scraped_data.add_review(Review(
        product_url=r.get('product_url', ''),
        product_name=r.get('product_name', ''),
        store_name=r.get('store_name', ''),
        reviewer_name=r.get('reviewer_name', 'Anonymous'),
        rating=r.get('rating', 5),
        review_text=r.get('review_text', ''),
        review_date=r.get('review_date'),
        variant=r.get('variant'),
        helpful_count=r.get('helpful_count', 0),
        review_id=r.get('review_id', '')
    ))

# FRESH SAVE - no loading existing IDs
dm = DataManager()
# dm.load_existing_ids()  # SKIP THIS for fresh start

# Deduplicate within this batch only
original = len(scraped_data.reviews)
seen = set()
unique = []
for r in scraped_data.reviews:
    if r.review_id not in seen:
        seen.add(r.review_id)
        unique.append(r)
scraped_data.reviews = unique
print(f"🔄 Deduplicated: {original} → {len(scraped_data.reviews)}")

# Save
output_paths = dm.save_all(scraped_data)
print(f"\n✅ Saved!")
print(f"📁 CSV: {output_paths['csv']}")
print(f"📁 JSON: {output_paths['json']}")
```

## 6️⃣ Final Statistics

```python
df = pd.read_csv(settings.OUTPUT_CSV)

print(f"📊 FINAL DATASET STATISTICS")
print(f"{'='*50}")
print(f"Total reviews: {len(df)}")
print(f"Unique products: {df['product_name'].nunique()}")
print(f"Unique reviewers: {df['reviewer_name'].nunique()}")
print(f"Average rating: {df['rating'].mean():.2f}")
print(f"Avg review length: {df['review_text'].str.len().mean():.0f} chars")

print(f"\n📈 Rating Distribution:")
print(df['rating'].value_counts().sort_index())

print(f"\n📦 Top 10 Products by Reviews:")
print(df.groupby('product_name').size().sort_values(ascending=False).head(10))

# Check if target reached
if len(df) >= 2000:
    print(f"\n🎉 TARGET REACHED! {len(df)} reviews collected!")
else:
    print(f"\n⚠️ Need {2000 - len(df)} more reviews to reach target")
```
